```python
B,G,R= cv2.split(image) #extracting B G R channels from an image (each channels are array stacked )
```

```python
cv2.imshow("Blue",B)     #
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("Green",G)     #
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("Red",R)     #
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
added=cv2.add(B,G)
cv2.imshow("Added",added)    
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
Weighted=cv2.addWeighted(B,0.5,G,0.5,0.2)
cv2.imshow("Added",weighted)    
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
weighted=cv2.addWeighted(B,0.5,G,0.5,0.2)
cv2.imshow("Added",weighted)    
cv2.waitKey(0)
cv2.destroyAllWindows()
```
